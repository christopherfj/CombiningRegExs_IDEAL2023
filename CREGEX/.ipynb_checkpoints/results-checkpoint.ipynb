# Classification results

```python
import pickle
from sklearn.metrics import accuracy_score, f1_score
import os
from collections import Counter, defaultdict
import pandas as pd
from collections import defaultdict
import numpy as np
import pandas as pd

from matplotlib import pylab
import seaborn as sns
sns.set_style("dark")
sns.set(font_scale=1.8) 

FILENAMES = ['OBESIDAD', 'OBESIDAD_TIPOS','FUMADOR'] 

TITLES = {
          'OBESIDAD': 'OBESITY STATUS',
          'OBESIDAD_TIPOS': 'OBESITY TYPES',
          'FUMADOR': 'SMOKING STATUS',
}

MODELS = [
        'cregex-rf-n1', 'rf-n1',
        'cregex-svm-n1', 'svm-n1',
        'cregex-nb-n1', 'nb-n1',
        'cregex-bert', 'bert',
        #'cregex-sw', 
        'cregex-random'
         ]

#PROBS = ['0.5', '0.8', '0.85', '0.9', '0.95', '0.99']
#PROBS = ['0.95']
#ANALYSIS = True
PROBS = ['0.95']
ANALYSIS = False

RUNS = 1
FOLDS = 5
distributions = defaultdict(list)
results = defaultdict(list)

if ANALYSIS:
    fig = pylab.figure(1, figsize=(15,8))
    
f = 0
for FILENAME in FILENAMES:
    deltas_acc = []
    deltas_f1 = []
    print('-'*30)
    print(FILENAME)
    f+=1
    p = 0
    for PROB in PROBS:
        print(PROB)
        BINS = np.linspace(float(PROB), 1.0, 10)
        SAVE = True
        p+=1
        for MODEL in MODELS:
            print(MODEL)
            f1 = 0
            acc = 0
            hyps = defaultdict(list)
            if 'cregex' in MODEL:
                PROB_AUX = '_'+PROB
            else:
                PROB_AUX = ''
            for r in range(RUNS):
                for k in range(FOLDS):
                    y_true = []
                    pred = []
                    hyps_aux = []
                    dstb = []
                    with open( os.path.join( os.getcwd(), 'out', 'RESULTS', FILENAME, FILENAME+'_results_'+MODEL+PROB_AUX+'_r'+str(r+1)+'_f'+str(k+1)+'.pkl' ), 'rb') as a:
                        y_true, pred, hyps_aux, __, dstb_aux = pickle.load(a)
                        if 'cregex' in MODEL:
                            if ANALYSIS:
                                dstb = np.array( [val for val,_ in dstb_aux['predict']] )
                                idxs = np.where(dstb=='rex')[0]
                                y_true = y_true[idxs]
                                pred = pred[idxs]
                                confs_fit = np.array( dstb_aux['fit_all'][1][2] ) 
                                confs_pred = np.array( [rlc[2][0] for val,rlc in dstb_aux['predict'] if val=='rex'] )
                                if SAVE:
                                    counts_fit, bins_fit = np.histogram(confs_fit, bins=BINS)
                                    counts_pred, bins_pred = np.histogram(confs_pred, bins=BINS)
                                    distributions[FILENAME+PROB_AUX].append( [bins_fit, counts_fit, bins_pred, counts_pred] )
                            else:
                                results[FILENAME+'_'+MODEL+PROB_AUX].append( [y_true, pred, dstb_aux] )

                        if 'bert' not in MODEL:
                            for key in hyps_aux:
                                hyps[key].append(hyps_aux[key])
                                                    
                        acc += accuracy_score(y_true, pred)*1/(RUNS*FOLDS)
                        f1 += f1_score(y_true, pred, average='weighted')*1/(RUNS*FOLDS)
                            
            print('ACC-F1' )
            print(round(100*acc,2), '&', round(100*f1,2) )
            if 'cregex' in MODEL:
                delta_acc = round(100*acc,2)
                delta_f1 = round(100*f1,2)
            else:
                delta_acc -= round(100*acc,2)
                delta_f1 -= round(100*f1,2)
                deltas_acc.append(delta_acc)
                deltas_f1.append(delta_f1)
                print('Delta ACC - Delta F1' )
                print(round(delta_acc,2), '&', round(delta_f1,2) )
                if 'bert' not in MODEL:
                    for key in hyps:
                        print( key, Counter(hyps[key]) )
            SAVE = False
            
        if ANALYSIS:
            y = np.array( [ d for _,__,___,d in distributions[FILENAME+'_'+PROB] ] )
            y = np.mean(y, axis=0, dtype=int)
            x = np.array( [ d for _,__,d,___ in distributions[FILENAME+'_'+PROB] ][0] )
            x = (x[:-1] + x[1:]) / 2
            
            dct = {'PR VALUES':x,'FREQUENCY':y}
            df = pd.DataFrame.from_dict(dct)
            MED = int(len(x)/2)
            xticks = ['']*len(x)
            xticks[0] = str(BINS[0])
            xticks[MED] = str(round(BINS[MED],2))
            xticks[-1] = str(BINS[-1])
            
            ax = pylab.subplot(len(PROBS),len(FILENAMES),len(PROBS)*(f-1)+p)

            axd = df.plot('PR VALUES', 'FREQUENCY', kind='bar', legend=False, ax=ax, xticks=df.index, rot=0 )
            axd.set_xticklabels(xticks)
            pylab.ylabel('FREQUENCY')
            pylab.grid(True)
            pylab.title(TITLES[FILENAME]+'\nPR='+str(PROB), fontweight='bold')
            
    print('acc:', np.array(deltas_acc).mean(), 'f1:', np.array(deltas_f1).mean())

if ANALYSIS:
    fig.tight_layout()
    pylab.show()
    
```

# Classification analysis

```python
from matplotlib import pylab
import os
from collections import defaultdict
from sklearn.metrics import accuracy_score, f1_score, zero_one_loss
import numpy as np
from matplotlib.patches import Patch
import seaborn as sns
sns.set_style("dark")
sns.set(font_scale=2.0) 

RUNS = 1
FOLDS = 5

FILENAMES = ['OBESIDAD', 'OBESIDAD_TIPOS','FUMADOR'] 
PROB = '0.95'

MODELS = [
        'cregex-rf-n1_', 
        'cregex-svm-n1_', 
        'cregex-nb-n1_', 
        'cregex-bert_', 
        #'cregex-sw_', 
        'cregex-random_'    
         ]
MODELS = [MODEL+PROB for MODEL in MODELS]

colors = ['g','b','r','y','c','m']
labels = ['RegExs&RF', 'RegExs&SVM', 'RegExs&NB', 'RegExs&BERT', 'RegExs&RAND']

TITLES = {
          'OBESIDAD': 'OBESITY STATUS',
          'OBESIDAD_TIPOS': 'OBESITY TYPES',
          'FUMADOR': 'SMOKING STATUS',
}

patches = [Patch(color=v, label=k) for k, v in dict(zip(labels, colors)).items()]
fig = pylab.figure(1, figsize=(15,8))

idx = 0

for FILENAME in FILENAMES:
    print(FILENAME)
    idx+=1
    y = []
    xticks = []
    matches = []
    
    for MODEL in MODELS:
        print(MODEL)
        error_rex = []
        error_clf = []
        y_true_aux = []
        pred_aux = []
        
        for d in range(RUNS*FOLDS):
            y_true_aux, pred_aux, dstb_aux = results[FILENAME+'_'+MODEL][d]
            dstb = np.array( [val for val,_ in dstb_aux['predict']] )
            
            idxs = np.where(dstb=='rex')[0]
            y_true = y_true_aux[idxs]
            pred = pred_aux[idxs]            
            
            if len(matches)<RUNS*FOLDS:
                matches.append(100*len(idxs)/len(y_true_aux))
            
            idxs = np.where(dstb=='clf')[0]
            y_true = y_true_aux[idxs]
            pred = pred_aux[idxs]
            error_clf.append( 100*zero_one_loss(y_true, pred) )
            
        y.append( np.mean(error_clf) )
        xticks.append('')

    x = np.arange(len(y))    
    pylab.subplot(1,3,idx)
    pylab.title('\n'+TITLES[FILENAME], fontweight='bold')
    pylab.xticks(x,xticks)
    pylab.bar(x,y, color = colors)
    pylab.ylabel('AVERAGE ZERO-ONE-LOSS (%)')
    pylab.xlabel('CLASSIFIERS')
    pylab.grid(True)
    pylab.ylim([0,100])
    print('bars', y)
    
    matches = np.array(matches).mean()
    print('matches:', matches)
    
fig.tight_layout()
pylab.legend(handles = patches, loc='upper center', bbox_to_anchor=(-1.0, -0.1), ncol=5)
pylab.show()
```

```python
import sys
sys.path.append('../FREGEX')
import warnings
warnings.filterwarnings("ignore")
import os
import pickle
from sklearn.metrics import accuracy_score, f1_score, zero_one_loss
import numpy as np
import matplotlib
from matplotlib import pylab
import seaborn as sns
import pandas as pd
from IPython.display import display
from tabulate import tabulate
from collections import Counter, defaultdict
import re
import copy
from scipy import stats  
from scipy.stats import wilcoxon
from utils import complete_value
sns.set_style("dark")
sns.set(font_scale=1.8) 

FILENAMES = ['OBESIDAD', 'OBESIDAD_TIPOS', 'FUMADOR']

MODELS = [
               'cregex-rf-n1',
               'cregex-svm-n1',           
               'cregex-nb-n1',
               'cregex-bert',
               #'cregex-random'
               ]
RUNS = 1
FOLDS = 5
BATCH = 64

LABELS_CLASSES = {
           'OBESIDAD': ['NEGATIVE', 'POSITIVE'], 
           'OBESIDAD_TIPOS':['MODERATE', 'SEVERE', 'MORBID'],
           'FUMADOR': ['NEGATIVE', 'POSITIVE'],
           'AMAZON': ['NEGATIVE', 'POSITIVE'],
           'IMDB': ['NEGATIVE', 'POSITIVE'],
           'YELP': ['NEGATIVE', 'POSITIVE']
           }

METRIC = 'ZERO-ONE-LOSS'

pylab.rcParams["legend.facecolor"] = "white"
pylab.rcParams['figure.facecolor'] = "white"

CURVES = dict( [ (MODEL, ['_results_lc_pl_', ]) for MODEL in MODELS] )
COLORS = {'RF': 'g', 'SVM':'b', 'NB':'r', 'BERT': 'y', 'RAND':'m'}

TITLES = {
          'OBESIDAD': 'OBESITY STATUS',
          'OBESIDAD_TIPOS': 'OBESITY TYPES',
          'FUMADOR': 'SMOKING STATUS',
}

print('RUNS:', RUNS, 'FOLDS:', FOLDS, '\n')

cformat  = 'c'*( len(CURVES)+1 )
idx = 0

fig = pylab.figure(1, figsize=(15,20))

for FILENAME in FILENAMES:
    idx +=1
    MAX_SIZE = 0
    pos = 0
    curves_all = {}
    for MODEL in MODELS:
        pos+=1
        for CURVE in CURVES[MODEL]:
            CURVE_AUX = CURVE.split('_')[3].upper()  
            y_tr_axis = []
            y_test_axis = []
            for r in range(RUNS):
                for k in range(FOLDS):
                    with open( os.path.join( os.getcwd(), 'out', 'RESULTSLC', CURVE_AUX, FILENAME, FILENAME+CURVE+MODEL+'_r'+str(r+1)+'_f'+str(k+1)+'.pkl' ), 'rb') as a:
                        results_lc = pickle.load(a)
                        #learning curves
                        y_tr_aux = []
                        y_test_aux = []
                        for _, x_axis, ypred in results_lc[CURVE_AUX]:
                            for ypredb in ypred:
                                y_tr_aux.append( 100*zero_one_loss(ypredb[0], ypredb[1]) )
                                y_test_aux.append( 100*zero_one_loss(ypredb[2], ypredb[3]) )
                        y_tr_axis.append(y_tr_aux)
                        y_test_axis.append(y_test_aux)
                        if len(y_tr_aux)>MAX_SIZE:
                            MAX_SIZE = len(y_tr_aux)

            y_tr_axis = complete_value(y_tr_axis, MAX_SIZE)
            y_test_axis = complete_value(y_test_axis, MAX_SIZE)

            error_y_tr_axis = np.array(y_tr_axis).std(axis=0)
            error_y_test_axis = np.array(y_test_axis).std(axis=0)
            y_tr_axis = np.array(y_tr_axis).mean(axis=0)
            y_test_axis = np.array(y_test_axis).mean(axis=0)
            pylab.subplot(len(FILENAMES), 1, idx)
            pylab.title(TITLES[FILENAME], fontweight='bold')

            label_aux = MODEL.replace('cregex','').replace('-n1','').replace('-','').replace('random','rand').upper() 
            color_aux = COLORS[label_aux] 
            pylab.plot(x_axis, y_tr_axis, color_aux, label='RegExs&'+label_aux+' (training)'.upper(), lw=4, ms=10)
            pylab.plot(x_axis, y_test_axis, color_aux, label='RegExs&'+label_aux+' (test)'.upper(), lw=4, ms=10, linestyle='--')
            pylab.fill_between(x_axis, y_tr_axis-error_y_tr_axis, y_tr_axis+error_y_tr_axis, facecolor=color_aux[0],  alpha=0.08)
            pylab.fill_between(x_axis, y_test_axis-error_y_test_axis, y_test_axis+error_y_test_axis, facecolor=color_aux[0],  alpha=0.08)
            pylab.grid(True)  
            pylab.xlabel('Number of training examples'.upper() )
            pylab.ylabel( ('Average '+METRIC).upper()+' (%)' )  
            #pylab.ylim([0, 100])
            pylab.ylim([0, 30])

pylab.legend( loc='upper center', bbox_to_anchor=(0.45, -0.2), shadow=False, ncol=2 )
fig.tight_layout()
pylab.show()
fig.savefig( os.path.join(os.getcwd(), 'out', 'Figures', 'curves.PNG'), bbox_inches='tight' )
```
